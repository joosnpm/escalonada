```python
import numpy as np

def print_matriz(matriz, mensagem):
        print(mensagem)
        print(matriz)
        print()

def matriz_escalonada(matriz):
    m, n = matriz.shape
    matriz_escalonada = np.copy(matriz)

    linha_atual = 0
    for j in range(n):  # Para cada coluna
        # Encontra a primeira linha com um elemento não nulo na coluna j
        linha_nao_nula = None
        for i in range(linha_atual, m):
            if matriz_escalonada[i, j] != 0:
                linha_nao_nula = i
                break

        if linha_nao_nula is not None:
            # Troca a linha atual com a linha não nula encontrada
            if linha_nao_nula != linha_atual:
                matriz_escalonada[[linha_atual, linha_nao_nula]] = matriz_escalonada[[linha_nao_nula, linha_atual]]
                print_matriz(matriz_escalonada, f"TROCANDO A LINHA {linha_atual+1} COM A LINHA {linha_nao_nula+1}")

            # Normaliza a linha atual (apenas mostra a operação realizada para criação do pivô)
            pivote = matriz_escalonada[linha_atual, j]
            if pivote != 1:
                print(f"MULTIPLICANDO CADA ELEMENTO DA {linha_atual+1} POR {1/pivote}")
                for k in range(n):
                    matriz_escalonada[linha_atual, k] *= 1 / pivote
                print_matriz(matriz_escalonada, f"TRANSFORMANDO LINHA {linha_atual+1}")

            # Zera os elementos abaixo da linha atual na coluna j
            for i in range(linha_atual + 1, m):
                fator = matriz_escalonada[i, j]
                if fator != 0:
                    print(f"ZERANDO ELEMENTO DA LINHA {i+1}, COLUNA {j+1}")
                    matriz_escalonada[i] -= fator * matriz_escalonada[linha_atual]
                    print_matriz(matriz_escalonada, f"MULTIPLICANDO {fator} VEZES A LINHA {linha_atual+1} E SOMANDO COM A LINHA {i+1}")

            # Zera os elementos acima da linha atual na coluna j
            for i in range(linha_atual):
                fator = matriz_escalonada[i, j]
                if fator != 0:
                    print(f"ZERANDO ELEMENTO DA LINHA {i+1}, COLUNA {j+1}")
                    matriz_escalonada[i] -= fator * matriz_escalonada[linha_atual]
                    print_matriz(matriz_escalonada, f"MULTIPLICANDO {fator} VEZES A LINHA {linha_atual+1} E SOMANDO COM A LINHA {i+1}")

            linha_atual += 1

    return matriz_escalonada

    # Função para ler uma matriz digitada
def ler_matriz():
    m = int(input("DIGITE O NÚMERO DE LINHAS DA MATRIZ: "))
    n = int(input("DIGITE O NÚMERO DE COLUNAS DA MATRIZ: "))
    matriz = np.empty((m, n), dtype=float)
    print("DIGITE OS ELEMENTOS DA MATRIZ LINHA POR LINHA:")
    for i in range(m):
        linha = list(map(float, input().split()))
        matriz[i] = linha
    return matriz

    # Leitura da matriz do usuário
matriz_original = ler_matriz()

    # Transformação da matriz em forma escalonada
print("MATRIZ ORIGINAL:")
print(matriz_original)
print()
matriz_escalonada = matriz_escalonada(matriz_original)
# mostra a matriz final
print("MATRIZ ESCALONADA:")
print(matriz_escalonada)
```
